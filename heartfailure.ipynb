```python
import pandas as pd
d=pd.read_csv('/content/heart_failure_clinical_records_dataset.csv')
print(d)
```

```python
import seaborn as sns
sns.boxplot(data=d,x='DEATH_EVENT')
```

```python
from matplotlib import pyplot as plt
fig, ax=plt.subplots(1,2)
sns.histplot(data=d ,x=d['anaemia'],hue=d['DEATH_EVENT'], kde=True ,ax=ax[0] )
sns.boxplot(x=d['diabetes'],y=d['smoking'] ,ax=ax[1])
```

```python
corr_matrix=d.corr()
print(corr_matrix)
```

```python
cov_matrix=pd.DataFrame.cov(d)
sns.heatmap(cov_matrix,annot=True,fmt='g')
plt.show
```
